```python
import pandas as pd
```

```python
df =pd.read_csv('/content/drive/MyDrive/Malayalam_offensive/Malayalam_offensive_data_Training-YT.csv',encoding='utf-8')
```

```python
df
```

```python
df['Labels'].value_counts()
```

```python
d = {'NOT' : 0,'OFF' : 1}
```

```python
df['Labels'] = df['Labels'].map(d)
```

```python
df['Labels']
```

```python
labels = list(df['Labels'])
```

```python
tweets = df['Tweets']
```

```python
tweets
```

```python
import re
```

```python
# regex_for_english_hindi_emojis="[^a-zA-Z#\U0001F300-\U0001F5FF'|'\U0001F600-\U0001F64F'|'\U0001F680-\U0001F6FF'|'\u2600-\u26FF\u2700-\u27BF\u0900-\u097F]"
def clean_tweet(tweet):
    tweet = re.sub(r"@[A-Za-z0-9]+",' ', tweet)
    tweet = re.sub(r"https?://[A-Za-z0-9./]+",' ', tweet)
    # tweet = re.sub(regex_for_english_hindi_emojis,' ', tweet)
    tweet = re.sub("RT ", " ", tweet)
    tweet = re.sub("\n", " ", tweet)
    tweet = re.sub(r" +", " ", tweet)
    # tokens = []
    # for token in tweet.split():
    #     if token not in stopword:
    #         token = english_stemmer.stem(token)
    #         token = hi_stem(token)
    #         tokens.append(token)
    return tweet
```

```python
cleaned_tweets = [clean_tweet(tweet) for tweet in tweets]
```

```python
cleaned_tweets 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(cleaned_tweets)
# X = X.todense()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(4)
classifier.fit(X_train, y_train)
```

```python
X_train
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = classifier.predict(X_val)
print("With K-Nearest Neighbour:")
print(classification_report(y_val, y_pred))
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LogisticRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_val)
print("With K-Nearest Neighbour:")
print(classification_report(y_val, y_pred))
```
